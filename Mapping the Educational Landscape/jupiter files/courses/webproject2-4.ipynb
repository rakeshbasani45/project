```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
colleges=[]
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/116.0.0.0 Safari/537.36 Edg/116.0.1938.54'}
for i in range(1,181):
    data = requests.get(f"https://engineering.careers360.com/colleges/list-of-engineering-colleges-in-india?page={i}&sort_by=4",headers = headers)
    soup = BeautifulSoup(data.text,"html.parser")
    college = soup.find("body")\
              .find("div",id="root")\
              .find("section",class_="college_listing_page")\
              .find("div",class_="undefined")\
              .find("div",class_="row")\
              .find("div",class_="desktop_right_side col-md-9")\
              .find_all("div",class_="card_block")
    
    for col in college:
            try:
                college_name = col.find("div",class_="tupple_right_block d-none d-md-block")\
                                  .find("h3",class_="college_name").text
                colleges.append(college_name)
            except Exception as e:
                print("na")
colleges
```

```python
import openpyxl
excelwb = openpyxl.Workbook()
sheet = excelwb.active
sheet.append(["college","course_name","duration","seats","fees","seats1","fees1"])
```

```python
for i in colleges:
    try:
        cleaned_college_name = i.lower().split(" - ")[1].replace(" ", "-").replace(",","")
        data = requests.get(f"https://www.careers360.com/colleges/{cleaned_college_name}/courses",headers = headers)
        soup = BeautifulSoup(data.text,"html.parser")
        course = soup.find("body")\
                     .find("div",id="root")\
                     .find("div",class_="undefined college_main_container")\
                     .find("section",class_="courses_fees pt-5")\
                     .find("div",class_="container")\
                     .find_all("div",class_="card_blk")
        for cour in course:
            try:
                college = cour.find("div",class_="detail")\
                              .find("div",class_="course_detail")\
                              .find("div",class_="accepted_exam offered_by").text.split(":")[1]
                course_name = cour.find("div",class_="detail")\
                                   .find("h4").text
                duration = cour.find("div",class_="detail")\
                               .find("span").text
                try:
                    total_seats = cour.find("div",class_="detail")\
                                      .find_all("span")[1].text
                    if total_seats.isdigit():
                        seats = (total_seats)
                        fees = ("none")
                    else:
                        seats = ("none")
                        fees =(total_seats)
                except:
                    seats = ("none")
                    fees = ("none")
                try:
                    total_fees = cour.find("div",class_="detail")\
                                    .find_all("span")[2].text
                    if total_fees.isdigit():
                        seats1 = (total_fees)
                        fees1 = ("none")
                    else:
                        seats1 = ("none")
                        fees1 =(total_fees)
                except:
                    seats1 = ("none")
                    fees1 = ("none")
                sheet.append([college,course_name,duration,seats,fees,seats1,fees1])
                print([college,course_name,duration,seats,seats1,fees1])
            except Exception as e:
                print("na")
            excelwb.save(r"D:\project 2\data_NA.xlsx")
    except Exception as e:
        print("NA")
```
