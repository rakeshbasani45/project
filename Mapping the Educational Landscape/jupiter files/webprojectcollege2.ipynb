```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
import openpyxl
excelwb = openpyxl.Workbook()
sheet = excelwb.active
sheet.append(["college_name","location","college_type","rating","total_reviews"])
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/116.0.0.0 Safari/537.36 Edg/116.0.1938.54'}
for i in range(1,181):
    data = requests.get(f"https://engineering.careers360.com/colleges/list-of-engineering-colleges-in-india?page={i}&sort_by=4",headers = headers)
    soup = BeautifulSoup(data.text,"html.parser")
    college = soup.find("body")\
              .find("div",id="root")\
              .find("section",class_="college_listing_page")\
              .find("div",class_="undefined")\
              .find("div",class_="row")\
              .find("div",class_="desktop_right_side col-md-9")\
              .find_all("div",class_="card_block")
    
    for col in college:
            try:
                college_name = col.find("div",class_="tupple_right_block d-none d-md-block")\
                                  .find("h3",class_="college_name").text
                Location = col.find("div",class_="tupple_right_block d-none d-md-block")\
                              .find_all("span")[0].text
                college_type = col.find("div",class_="tupple_right_block d-none d-md-block")\
                                  .find_all("span")[1].text
                try:
                    rating = col.find("div",class_="tupple_right_block d-none d-md-block")\
                                .find_all("span")[2].text
                except:
                    rating ="Not Mention"
                try:
                    total_reviews =col.find("div",class_="tupple_right_block d-none d-md-block")\
                                      .find_all("span")[3].text
                except:
                    total_reviews = "Not Mention"
                
                sheet.append([college_name,Location,college_type,rating,total_reviews])
                print([college_name,Location,college_type,rating,total_reviews])

            except Exception as e:
                print("na")
            excelwb.save(r"D:\college2.xlsx")
```
