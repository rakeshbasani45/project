```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
import openpyxl
excelwb = openpyxl.Workbook()
sheet = excelwb.active
sheet.append(["College_name","available_facilities"])
```

```python
colleges=[]
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/116.0.0.0 Safari/537.36 Edg/116.0.1938.54'}
for i in range(1,181):
    data = requests.get(f"https://engineering.careers360.com/colleges/list-of-engineering-colleges-in-india?page={i}&sort_by=4",headers = headers)
    soup = BeautifulSoup(data.text,"html.parser")
    college = soup.find("body")\
              .find("div",id="root")\
              .find("section",class_="college_listing_page")\
              .find("div",class_="undefined")\
              .find("div",class_="row")\
              .find("div",class_="desktop_right_side col-md-9")\
              .find_all("div",class_="card_block")
    for col in college:
            try:
                college_name = col.find("div",class_="tupple_right_block d-none d-md-block")\
                                  .find("h3",class_="college_name").text
                colleges.append(college_name)
            except Exception as e:
                print("na")
colleges
```

```python
for i in colleges:
    try:
        college=i.lower().replace(" ", "-").replace(",", "")
        data = requests.get(f"https://www.careers360.com/colleges/{college}/facilities",headers = headers)
        soup = BeautifulSoup(data.text,"html.parser")
        Fac=soup.find("body")\
                .find("div",id="root")\
                .find("div",class_="college_main_container")\
                .find("section",class_="pt-5")\
                .find("div",class_="container")\
                .find("div",class_="row")\
                .find("div",class_="facilities_list")\
                .find("ul")\
                .find_all("li")

        for fac in Fac:
            try:
                College_name = college
                available_facilities = fac.find("div",class_="facilities_wrap")\
                                          .find_all("span")[2].text
                sheet.append([College_name,available_facilities])
                print([College_name,available_facilities])
            except Exception as e:
                print("na")      
            excelwb.save(r"D:\project 2\datafacilities.xlsx")
    except Exception as e:
        print("NA")
```
